## **1. Marco del problema:**

### **1.1. Objetivo del proyecto:**

El objetivo de este proyecto es desarrollar un modelo de predicción automática de fallos en distintas máquinas que permita predecir la ocurrencia de posibles averías antes de que ocurran.

Esto:
  - Reducirá el tiempo de inactividad de las máquinas al no produrcirse paradas de los equipos no planificadas para la realización de mantenimientos correctivos.
  - Reducirá la producción de material defectuoso como consecuencia de los fallos en los equipos.
  - Optimizará los trabajos de mantenimiento, ya que podrá realizarse un mantenimiento preventivo únicamente cuando sea necesario, en base a la predicción del posible fallo, antes de que ocurra.
  - Aumentará la productividad de los equipos debido a:
    - La reducción de paradas innecesarias.
    - La disminución de las piezas defectuosas.





### **1.2.Tipo de problema:**

Este es un problema de **aprendizaje supervisado**, ya que el conjunto de datos incluye etiquetas que indican la presencia o ausencia de fallos en las máquinas.

Se trata de un problema de **clasificación** ya que se pretende clasificar si una máquina va a fallar o no en función de los valores obtenidos para diferentes sensores.

### **1.4. Descripción del dataset:**

Este conjunto de datos contiene datos de sensores recogidos de varias máquinas.

Incluye diversas lecturas de sensores, así como los fallos registrados de las máquinas.

Descripción de las columnas:

- **footfall:**
  - El número de personas u objetos que pasan por la máquina. (footfall: the number of people entering a shop or shopping area in a given time)
  - Puede influir en el desgaste o acumulación de partículas, afectando el rendimiento.
  - Los valores en el csv van de 0 (min) hasta 7300 (max), y el promedio media es 306,38 (sin eliminar las filas duplicadas).

- **tempMode:**
  - El modo o ajuste de temperatura en el que opera la máquina.
  - Puede estar relacionado con diferentes condiciones de trabajo.
  - Los valores posibles son números enteros de 0 a 7.
  - Los valores en el csv van de 0 (min) hasta 7 (max), y el promedio media es 3,72.

- **AQ:**
  - Índice de calidad (pureza) del aire cerca de la máquina.
  - Una mala calidad del aire (presencia de polvo, humedad, partículas) puede contribuir al deterioro o mal funcionamiento.
  - Los valores posibles son números enteros de 1 (calidad excelente) a 7 (calidad muy mala).

- **USS:**
  - Datos del sensor ultrasónico o Ultrasonic Sensors, que indica las mediciones de proximidad a la máquina.
  - Puede indicar la proximidad de materiales, obstáculos o acumulaciones de residuos que podrían afectar el funcionamiento.

- **CS:**
  - Lecturas del sensor de corriente, indicando el uso de corriente eléctrica de la máquina, es decir, el consumo eléctrico de la máquina.
  - Un aumento inesperado de corriente puede indicar sobrecarga o fallo inminente en componentes eléctricos.
  - Los valores posibles son números enteros de 1 (muy poca corriente eléctrica) a 7 (máxima corriente eléctrica).

- **VOC:**
  - Nivel de compuestos orgánicos volátiles detectado cerca de la máquina, es decir, la concentración de sustancias químicas en el aire que se vaporizan a temperatura ambiente..
  - La exposición a altos niveles de VOC puede afectar la vida útil de los componentes internos.
  - Los valores posibles van de 0 (nivel nulo) hasta 6 (nivel muy alto).

- **RP:**
  - Posición rotacional o RPM (Revoluciones Por Minuto) de las piezas de la máquina.
  - Cambios anómalos en la velocidad pueden ser señales de desgaste mecánico o problemas en la lubricación.
  - Los valores en el csv van de 19 (min) hasta 91 (max), y el promedio media es 47.

- **IP:**
  - Presión de entrada del fluido (aire, agua, aceite, etc.) a la máquina o Input Pressure.
  - Bajadas de presión podrían indicar fugas, bloqueos o problemas en la alimentación del sistema.
  - Los valores posibles van de 1 (presión muy baja) hasta 7 (presión muy alta).

- **Temperature:**
  - Temperatura de funcionamiento de la máquina.
  - Un aumento repentino podría señalar fricción excesiva, sobrecarga o fallos en la refrigeración.
  - Los valores en el csv van de 1 (min) hasta 24 (max), y el promedio media es 16,33.

- **fail:**
  - Indicador binario de fallo de la máquina (1 para fallo, 0 para ningún fallo).
  - Variable objetivo a predecir.

### **1.6. Medición del rendimiento:**

Para evaluar la capacidad predictiva del modelo de clasificación en este proyecto, se emplearán diversas métricas que nos permitirán cuantificar su precisión y efectividad al determinar si una máquina fallará o no, basándonos en los parámetros disponibles. Estas métricas clave incluyen:

- **Exactitud (Accuracy)**: Representa el porcentaje de predicciones correctas realizadas por el modelo sobre el total de predicciones. Se calcula como [\frac{\text{Número de predicciones correctas}}{\text{Número total de predicciones}}]. Una exactitud alta indica un buen rendimiento general del modelo.

- **Precisión (Precision)**: Mide la proporción de instancias clasificadas como positivas (fallo de la máquina) que realmente lo fueron. Se calcula como [\frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos + Falsos Positivos}}]. Una alta precisión indica que cuando el modelo predice un fallo, es muy probable que realmente ocurra.

- **Recall (Sensibilidad o Exhaustividad)**: Mide la proporción de instancias positivas reales (fallos de la máquina) que fueron correctamente identificadas por el modelo. Se calcula como [\frac{\text{Verdaderos Positivos}}{\text{Verdaderos Positivos + Falsos Negativos}}]. Un alto recall indica que el modelo es bueno para detectar la mayoría de los fallos reales.

- **F1-Score**: Es la media armónica de la precisión y el recall, proporcionando una medida equilibrada del rendimiento del modelo, especialmente útil cuando las clases están desbalanceadas. Se calcula como [2 \times \frac{\text{Precisión} \times \text{Recall}}{\text{Precisión + Recall}}].

- **Curva ROC y AUC (Área Bajo la Curva ROC)**: La curva ROC grafica la tasa de verdaderos positivos (TPR) contra la tasa de falsos positivos (FPR) para diferentes umbrales de clasificación. El AUC mide el área bajo esta curva, representando la capacidad del modelo para distinguir entre la clase positiva (fallo) y la clase negativa (no fallo). Un AUC cercano a 1 indica un excelente rendimiento.

- **Validación Cruzada**: Se implementará una estrategia de validación cruzada para asegurar que el modelo no solo funcione bien con los datos de entrenamiento, sino que también generalice de manera efectiva a datos nuevos e imprevistos. Esto ayuda a mitigar el riesgo de sobreajuste y a obtener una evaluación más robusta del rendimiento real del modelo.

Estas métricas en conjunto ofrecerán una evaluación exhaustiva de la capacidad del modelo de regresión logística para predecir con precisión la probabilidad de fallo de la máquina, permitiendo la optimización y el ajuste fino para lograr el mejor rendimiento posible.

## **6. Creación de Pipelines de Preprocesamiento y Modelado:**

El siguiente paso en el proyecto es la creación de pipelines de preprocesamiento y modelado.
La implementación de pipelines en scikit-learn ofrece varios beneficios fundamentales:

- **Organización y Claridad:** Permiten estructurar de manera lógica y secuencial los pasos de preprocesamiento y el modelo de entrenamiento, lo que facilita la comprensión y el mantenimiento del código.
- **Prevención de Fuga de Información (Data Leakage):** Al aplicar el preprocesamiento dentro del pipeline y durante la validación cruzada, aseguramos que las transformaciones se aprendan únicamente de los datos de entrenamiento de cada partición, evitando que información del conjunto de validación o prueba influya en el proceso de entrenamiento.
- **Reproducibilidad:** Garantizan que el mismo flujo de trabajo se aplique de manera consistente a los datos de entrenamiento, validación y prueba, así como a futuros datos de predicción.
- **Simplificación del Flujo de Trabajo:** Una vez definidos, los pipelines permiten entrenar el modelo y realizar predicciones con una sola llamada a los métodos `fit` y `predict`.

El problema para el que se entrenarán los modelos de machine learning es un problema de clasificación binaria, ya que la variable objetivo toma dos posibles valores: 0 (ausencia de fallo) y 1 (presencia de fallo).

Para abordar este problema, se entrenarán distintos modelos de clasificación que son adecuados para este tipo de tarea:

- **Modelos Lineales:**

  - **LogisticRegression**: Modelo lineal que estima la probabilidad de una clase en función de una combinación lineal de las características de entrada. Es comúnmente utilizado para problemas de clasificación binaria y multiclase.

- **Modelos Basados en Árboles:**

  - **DecisionTreeClassifier**: Modelo basado en árboles de decisión que divide los datos en ramas según las decisiones tomadas en cada nodo, basadas en los valores de las características. Es intuitivo y puede capturar relaciones no lineales.
  - **RandomForestClassifier**: Conjunto de múltiples árboles de decisión entrenados en subconjuntos aleatorios de los datos y utilizando subconjuntos aleatorios de las características. Las predicciones se realizan por votación de los árboles, lo que mejora la generalización y reduce el sobreajuste.
  - **GradientBoostingClassifier**: Algoritmo de boosting que construye un modelo aditivo secuencialmente, donde cada nuevo modelo débil (típicamente árboles de decisión) corrige los errores del modelo anterior. Optimiza la clasificación combinando múltiples modelos débiles en un modelo fuerte.
  - **LightGBM**: Un framework de boosting de gradiente diseñado para ser distribuido y altamente eficiente. Utiliza técnicas como el muestreo de gradiente basado en histogramas y el crecimiento de hojas por niveles para acelerar el entrenamiento y reducir el uso de memoria.
  - **XGBoost**: Una implementación popular y eficiente del algoritmo de boosting de gradiente. Proporciona regularización L1 y L2, manejo de valores faltantes y paralelización, lo que lo hace potente y robusto.

- **Modelos Basados en Distancia:**

  - **KNeighborsClassifier**: Clasificador basado en la proximidad de los puntos a sus vecinos más cercanos en el espacio de características. La clase de un nuevo punto se decide por la mayoría de las clases de sus k vecinos más cercanos.
  - **SVC (Support Vector Classifier)**: Modelo basado en máquinas de soporte vectorial, que encuentra un hiperplano óptimo en un espacio de alta dimensión para separar las clases, maximizando el margen entre ellas. Puede utilizar diferentes funciones kernel para modelar relaciones no lineales.

- **Modelos Probabilísticos**:

  - **GaussianNB**: Modelo basado en la teoría de Bayes, que asume que las características de cada clase siguen una distribución gaussiana (normal). Es un clasificador probabilístico que calcula la probabilidad de pertenencia a cada clase basándose en las probabilidades condicionales de las características.

Dado que voy a entrenar varios tipos de modelos de clasificación, cada uno con diferentes sensibilidades al preprocesamiento, he decidido crear tres pipelines distintos para optimizar el rendimiento de cada grupo de modelos.

- **Pipeline 1**: Transformación Logarítmica + Normalización + Balanceo.
  
  Este pipeline se centrará en aplicar una transformación logarítmica para reducir la asimetría de ciertas variables, normalizar todas las variables numéricas para que tengan una escala similar (crucial para modelos lineales y basados en distancia) y aplicar una técnica de balanceo para abordar el posible desequilibrio en la variable objetivo.

    - Pasos del Pipeline:
        1.  Aplicación de la transformación logarítmica a las variables identificadas con skewness (beneficioso para modelos lineales y probabilísticos).
        2.  Escalado de todas las variables numéricas utilizando 'StandardScaler' (esencial para modelos lineales y basados en distancia).
        3.  Aplicación de una técnica de balanceo de la variable objetivo utilizando 'SMOTE' en el conjunto de entrenamiento.
        4.  Selección de características.
        5.  Entrenamiento de los modelos.

    - Modelos a Entrenar con este Pipeline:
        1. LogisticRegression
        2. SVC (Support Vector Classifier)
        3. KNeighborsClassifier
        4. GaussianNB

- **Pipeline 2**: Transformación Logarítmica + Balanceo.

  Este pipeline se centrará en aplicar la transformación logarítmica para reducir la asimetría de ciertas variables y aplicar una técnica de balanceo de la variable objetivo, sin incluir la normalización (escalado), ya que los modelos basados en árboles son generalmente menos sensibles a la escala de las características.

    - Pasos del Pipeline:
        1.  Aplicación de la transformación logarítmica a las variables identificadas con skewness.
        2.  Aplicación de una técnica de balanceo de la variable objetivo utilizando 'SMOTE' en el conjunto de entrenamiento.
        3.  Selección de características.
        5.  Entrenamiento de los modelos.

    - Modelos a Entrenar con este Pipeline:
        1. DecisionTreeClassifier
        2. RandomForestClassifier
        3. GradientBoostingClassifier
        4. LightGBM
        5. XGBoost
        6. GaussianNB (para comparar el impacto del escalado)

- **Pipeline 3**: Solo Balanceo de la variable objetivo.

  Este pipeline será el más simple y solo aplicará la técnica de balanceo de la variable objetivo. Se probará con modelos basados en árboles para ver si el preprocesamiento adicional (transformación logarítmica y normalización) es realmente necesario para su rendimiento en este conjunto de datos.

    - Pasos del Pipeline:
        1.  Aplicación de una técnica de balanceo de la variable objetivo utilizando 'SMOTE' en el conjunto de entrenamiento.
        2.  Selección de características.
        3.  Entrenamiento de los modelos.

    - Modelos a Entrenar con este Pipeline:
        1. DecisionTreeClassifier
        2. RandomForestClassifier
        3. GradientBoostingClassifier
        4. LightGBM
        5. XGBoost

La implementación de estos pipelines permitirá experimentar de manera organizada con diferentes estrategias de preprocesamiento para cada tipo de modelo y seleccionar la combinación óptima que maximice el rendimiento predictivo de nuestro sistema de clasificación.

### **6.1. Pipeline 1:** Transformación Logarítmica + Normalización + Balanceo

```python
from imblearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import PowerTransformer, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score, roc_auc_score
from sklearn.feature_selection import SelectKBest, f_classif
import numpy as np

# Definir las columnas con skewness asimetria desviación y las columnas numéricas
cols_to_log = ['footfall','USS', 'RP']
numeric_cols = ['tempMode', 'AQ', 'CS', 'VOC', 'IP', 'Temperature']

# Crear el preprocesamiento para el Pipeline 1
preprocessor_pipeline_1 = ColumnTransformer(
    transformers=[
        ('log', PowerTransformer(method='yeo-johnson'), cols_to_log),
        ('scaler', StandardScaler(), numeric_cols)],
    remainder='passthrough')

# Definir los modelos a entrenar en este pipeline
models_pipeline_1 = {
    'LogisticRegression': LogisticRegression(random_state=42, solver='liblinear', class_weight='balanced'),
    'SVC': SVC(random_state=42, class_weight='balanced', probability=True),
    'KNeighborsClassifier': KNeighborsClassifier(n_neighbors=5),
    'GaussianNB': GaussianNB()
}

# Definir el espacio de hiperparámetros para cada modelo
param_grid_pipeline_1 = {
    'LogisticRegression': {
        'C': np.logspace(-3, 3, 7),
        'penalty': ['l1', 'l2']
    },
    'SVC': {
        'C': np.logspace(-3, 3, 7),
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    },
    'KNeighborsClassifier': {
        'n_neighbors': range(3, 11),
        'weights': ['uniform', 'distance']
    },
    'GaussianNB': {}
}

# Métricas de evaluación
scoring = {
    'accuracy': 'accuracy',
    'precision': 'precision',
    'f1': 'f1',
    'recall': 'recall',
    'roc_auc': 'roc_auc'
}

# Configuración de la validación cruzada estratificada
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Diccionario para almacenar los resultados
results_pipeline_1 = {}

# Iterar sobre los modelos
for name, model in models_pipeline_1.items():
    print(f"Entrenando y ajustando hiperparámetros para: {name}")

    # Crear el pipeline completo
    pipeline = Pipeline(steps=[('preprocessor', preprocessor_pipeline_1),
                           ('balancer', SMOTE(random_state=42)),
                           ('feature_selection', SelectKBest(score_func=f_classif)),
                           ('classifier', model)])

    # Definir el grid de hiperparámetros específico para el modelo
    param_grid = {f'classifier__{k}': v for k, v in param_grid_pipeline_1[name].items()}

    # Realizar la búsqueda de hiperparámetros con GridSearchCV
    grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring=scoring, refit='accuracy', n_jobs=-1)
    grid_search.fit(X_train_val, y_train_val)

    # Almacenar los resultados del mejor modelo
    results_pipeline_1[name] = {
        'best_score': grid_search.best_score_,
        'best_params': grid_search.best_params_,
        'best_estimator': grid_search.best_estimator_,
        'cv_results': grid_search.cv_results_
    }

    print(f"Mejor accuracy-score para {name}: {results_pipeline_1[name]['best_score']:.4f}")
    print(f"Mejores hiperparámetros para {name}: {results_pipeline_1[name]['best_params']}")
    print("-" * 50)


print("\nResultados detallados por métrica del mejor modelo en Pipeline 1:")

for name, results in results_pipeline_1.items():
    print(f"\nModelo: {name}")
    print(f"Mejor accuracy-score: {results['best_score']:.4f}")
    print(f"Mejores Hiperparámetros: {results['best_params']}")

    best_cv_results_df = pd.DataFrame(results['cv_results']).loc[results['cv_results']['rank_test_f1'] == 1]

    if not best_cv_results_df.empty:
        print(f"Accuracy (CV): {best_cv_results_df['mean_test_accuracy'].iloc[0]:.4f}")
        print(f"Precision (CV): {best_cv_results_df['mean_test_precision'].iloc[0]:.4f}")
        print(f"Recall (CV): {best_cv_results_df['mean_test_recall'].iloc[0]:.4f}")
        print(f"ROC AUC (CV): {best_cv_results_df['mean_test_roc_auc'].iloc[0]:.4f}")
    else:
        print("No se encontraron resultados detallados de validación cruzada para el mejor modelo.")
```

```python
from sklearn.metrics import confusion_matrix

# Columnas para el preprocesamiento
cols_to_log = ['footfall', 'USS', 'RP']
numeric_cols = ['tempMode', 'AQ', 'CS', 'VOC', 'IP', 'Temperature']

# Preprocesador
preprocessor_pipeline_1_refined = ColumnTransformer(
    transformers=[
        ('log', PowerTransformer(method='yeo-johnson'), cols_to_log),
        ('scaler', StandardScaler(), numeric_cols)
    ],
    remainder='passthrough'
)

# Modelo y sus hiperparámetros
model = LogisticRegression(
    random_state=42,
    solver='liblinear',
    class_weight='balanced',
    C=0.1,
    penalty='l2'
)

# Pipeline completo
pipeline_1 = Pipeline(steps=[
    ('preprocessor', preprocessor_pipeline_1_refined),
    ('balancer', SMOTE(random_state=42)),
    ('feature_selection', SelectKBest(score_func=f_classif)),
    ('classifier', model)
])

# Entrenar el modelo con los datos de entrenamiento
pipeline_1.fit(X_train, y_train)

# Hacer predicciones con el conjunto de test
y_pred = pipeline_1.predict(X_test)
```

```python
print(y_pred)
```

```python
# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)

# Graficar la matriz de confusión normalizada
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix / np.sum(conf_matrix), annot=True, fmt=".2%", cmap="Blues")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de Confusión Normalizada")
plt.show()
```
